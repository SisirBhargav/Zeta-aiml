```python
user_movie_df=df.pivot(index='userId',columns='movie',values='rating').reset_index(drop=True)
user_movie_df
```

```python
user_movie_df.fillna(0,inplace=True)
user_movie_df
```

```python
#calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
from scipy.spatial.distance import correlation

vec1=[5,4,0,0]
vec2=[4,5,3,0]
correlation(vec1,vec2)
```

```python
user_sim=1-pairwise_distances(user_movie_df.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index=df.userId.unique()
user_sim_df
```
