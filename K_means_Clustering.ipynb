```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path='/content/drive/MyDrive/Colab Notebooks/income.csv'
df=pd.read_csv(file_path)
df
```

```python
#min max scaler
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
#scatterplot
plt.scatter(df['Income($)'],df.Age)
plt.ylabel('Age')
plt.xlabel('Income($)')
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Income($)','Age']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1 = df[df['cluster'] == 0]
df2 = df[df['cluster'] == 1]
df3 = df[df['cluster'] == 2]
print(df1)
print(df2)
print(df3)
```

```python
#scatterplot
plt.scatter(df1['Income($)'],df1.Age,color='green')
plt.scatter(df2['Income($)'],df2.Age,color='red')
plt.scatter(df3['Income($)'],df3.Age,color='blue')
plt.xlabel('Income($)')
plt.ylabel('Age')
```
