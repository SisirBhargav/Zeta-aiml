ceate a dictionary with column name as key values in that add rows to predict marks of student

```python
# Define the dataset as a dictionary
student_data = {
    "Name": [],        # Student names
    "Age": [],         # Student age
    "Study_Hours": [], # Number of hours spent studying
    "Attendance": [],  # Attendance percentage
    "Previous_Marks": [], # Marks in the previous test
    "Predicted_Marks": [] # Marks predicted (can be calculated later)
}

# Adding a new row (a student's data)
def add_student(data, name, age, study_hours, attendance, prev_marks, predicted_marks=None):
    data["Name"].append(name)
    data["Age"].append(age)
    data["Study_Hours"].append(study_hours)
    data["Attendance"].append(attendance)
    data["Previous_Marks"].append(prev_marks)
    data["Predicted_Marks"].append(predicted_marks)

# Add rows
add_student(student_data, "Alice", 16, 5, 95, 88)
add_student(student_data, "Bob", 17, 3, 85, 75)
add_student(student_data, "Charlie", 16, 4, 90, 80)

# Example: Predict marks using a simple formula
def predict_marks(data):
    for i in range(len(data["Name"])):
        # Simple prediction formula (customize as needed)
        study_effect = data["Study_Hours"][i] * 5
        attendance_effect = data["Attendance"][i] * 0.1
        prev_effect = data["Previous_Marks"][i] * 0.7
        data["Predicted_Marks"][i] = study_effect + attendance_effect + prev_effect

# Predict marks for all students
predict_marks(student_data)

# Display the data
for i in range(len(student_data["Name"])):
    print({key: student_data[key][i] for key in student_data})
```
