```python
# Define the dataset as a dictionary
student_data = {
    "Name": [],        # Student names
    "Age": [],         # Student age
    "Study_Hours": [], # Number of hours spent studying
    "Attendance": [],  # Attendance percentage
    "Previous_Marks": [], # Marks in the previous test
    "Predicted_Marks": [] # Marks predicted (can be calculated later)
}

# Adding a new row (a student's data)
def add_student(data, name, age, study_hours, attendance, prev_marks, predicted_marks=None):
    data["Name"].append(name)
    data["Age"].append(age)
    data["Study_Hours"].append(study_hours)
    data["Attendance"].append(attendance)
    data["Previous_Marks"].append(prev_marks)
    data["Predicted_Marks"].append(predicted_marks)

# Add rows
add_student(student_data, "Alice", 16, 5, 95, 88)
add_student(student_data, "Bob", 17, 3, 85, 75)
add_student(student_data, "Charlie", 16, 4, 90, 80)

# Example: Predict marks using a simple formula
def predict_marks(data):
    for i in range(len(data["Name"])):
        # Simple prediction formula (customize as needed)
        study_effect = data["Study_Hours"][i] * 5
        attendance_effect = data["Attendance"][i] * 0.1
        prev_effect = data["Previous_Marks"][i] * 0.7
        data["Predicted_Marks"][i] = study_effect + attendance_effect + prev_effect

# Predict marks for all students
predict_marks(student_data)

# Display the data
for i in range(len(student_data["Name"])):
    print({key: student_data[key][i] for key in student_data})
```

```python
# data cleaning with imputations
import pandas as pd
import numpy as np

#sample datagram with missing values for helmet detection use case
dat= {
    'Age':[25,None,30,35,None,4.5],
    'Height':[5.5,6.1,5.9,None,5.8,5.2],
    'HelmetStatus':[1,None,1,0,1,1],
    'Gender':['M','F','M',None,'F','M'],
    'Experience':[2,4,6,1,None,None]

    }
df=pd.DataFrame(dat)
print("Original dataframe with missing valuer")
print(df)

df_filled_mean=df.copy()
df_filled_median=df.copy()
df_filled_mode=df.copy()

#replace missing values with mean
df_filled_mean['Age'].fillna(df['Age'].mean(),inplace=True)
df_filled_mean['Height'].fillna(df['Height'].mean(),inplace=True)
df_filled_mean['Experience'].fillna(df['Experience'].mean(),inplace=True)

#replace missing values with median
df_filled_median['Age'].fillna(df['Age'].median(),inplace=True)
df_filled_median['Height'].fillna(df['Height'].median(),inplace=True)
df_filled_median['Experience'].fillna(df['Experience'].median(),inplace=True)

#replace missing values with mode
df_filled_mode['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df_filled_mode['HelmetStatus'].fillna(df['HelmetStatus'].mode()[0],inplace=True)

#show results for all imputations
print("\nDataFrame with missing values replaced with mean:")
print(df_filled_mean)
print("\nDataFrame with missing values replaced with median:")
print(df_filled_median)
print("\nDataFrame with missing values replaced with mode:")
print(df_filled_mode)
```

```python
import numpy as np

#sample data

data = [10,12,14,13,100,87,5,15,16,14,13,11]

#step 1:calculate q1 (25th percentile) and Q3 (75th percent
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)

#step 2:calculate IQR
IQR=Q3-Q1

#step 3:determine the lower and upper bunds for outliers
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

#step 4:identify outliers
outliers=[x for x in data if x<lower_bound or x>upper_bound]

#step 5:print
print(f"Q1(25th percentile): {Q1}")
print(f"Q3(75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample dataset
np.random.seed(1555)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10, 1, 10)])
})

# Calculate the IQR for box plot-based outlier detection
Q1 = data['value'].quantile(0.25)  # Corrected to 'quantile'
Q3 = data['value'].quantile(0.75)  # Corrected to 'quantile'
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]

# Print outliers
print("Outliers:")
print(outliers)

# Visualize outliers using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['value'])
plt.title('Box Plot Showing Outliers')
plt.xlabel('Value')
plt.show()
```

```python
#normalisation with min max scaler
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#sample dataset
data={
    'age':[25,30,35,40,45],
    'height':[150,160,170,180,190],
    'weight':[50,60,70,80,90]
}

df=pd.DataFrame(data)
print("Original dataframe")
print(df)

#initialise min max scalar
scalar=MinMaxScaler()

#normalise the dataset
normalised_data=scalar.fit_transform(df)

normalised_df=pd.DataFrame(normalised_data,columns=df.columns)
print("Normalised dataframe")
print(normalised_df)

```
